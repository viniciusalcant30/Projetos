```python
# Vector field

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


# posi√ß√µes x e y no plano 
x,y = np.meshgrid(np.linspace(-5,5,20),np.linspace(-5,5,20)) #x e y representa a localiza√ß√£o no espa√ßo 
  
# derivadas
dy = x 
dx = np.ones(dy.shape)  

plt.figure(figsize=(8, 8))
plt.title("Vector Field: y'=x")
plt.quiver(x,y,dx,dy)
plt.show()
```

```python
# posi√ß√µes x e y no plano 
x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10)) #x e y representa a localiza√ß√£o no espa√ßo 
  
# derivadas
dy = x 
dx = np.ones(dy.shape) 


plt.figure(figsize=(10, 10))
plt.streamplot(x, y, dx, dy, density=1.4, linewidth=None, color='#A23BEC')
plt.title("Vector Field: y'=x")
  

plt.grid()
plt.show()
```
